```python
import sagemaker
from sagemaker.xgboost import XGBoost
from sagemaker.session import Session

# Initialize SageMaker session
sagemaker_session = sagemaker.Session()
role = sagemaker.get_execution_role()  # Uses the IAM role you created earlier
bucket = "mlopsproject3"  
```

```python
# Paths to processed data in S3
train_data = f"s3://mlopsproject3/processed/X_train.csv"
test_data = f"s3://mlopsproject3/processed/X_test.csv"
output_path = f"s3://mlopsproject3/models/"  # Where the model will be saved
```

```python
xgb_estimator = XGBoost(
    entry_point="model_training.py",  # Correct parameter name
    dependencies=['requirements.txt'],
    role=role,
    instance_count=1,
    instance_type="ml.m5.large",
    framework_version="1.7-1",
    output_path=output_path,
    hyperparameters={
        "objective": "binary:logistic",
        "eval_metric": "logloss",
        "max_depth": 5,
        "subsample": 0.8,
        "n_estimators": 100
    }
)
```

```python
xgb_estimator.fit({"train": train_data, "test": test_data})
```
